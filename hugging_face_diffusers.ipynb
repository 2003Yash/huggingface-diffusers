```python
from diffusers import DiffusionPipeline
```

```python
model_id = "CompVis/stable-diffusion-v1-4"

pipe = DiffusionPipeline.from_pretrained(model_id)
```

```python
prompt = "a photo of indian ninja"

images = pipe([prompt], num_inference_steps= 1, eta = 0.3, guidance_rescale=6).images # inference steps are no.of denoising steps of diffusion model
                                                                                      #  eta parameter is a hyperparameter that controls the amount of randomness during the denoising process
                                                                                      # guidance rescale is a parameter used to adjust the strength of the text prompt's influence on the generated image.
```

```python
plt.imshow(images[0])
```

```python
prompt = "a photo of indian ninja"

images = pipe([prompt], num_inference_steps= 50, eta = 0.3, guidance_rescale=6)["sample"]
```

```python
plt.imshow(images[0])
```
